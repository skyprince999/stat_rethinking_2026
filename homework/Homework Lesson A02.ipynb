Question: 

A2. Suppose the globe tossing data (Chapter 2) had turned out to be 3 water and 11 land. Construct the posterior distribution. Then using the posterior distribution, compute the  posterior predictive distribution for the next 5 tosses of the same globe.

----

Assumptions:

```
W (water) = 3
L (land) = 11
N (Total tosses) = 14 (W+L)
```

posterior distribution `p = P(water)`

Assuming _prior_ on `p` is *Uniform(0,1)*




````python
import numpy as np
from scipy.stats import beta, binom
import matplotlib.pyplot as plt
````

````python
# Observed data
W, L = 3, 11

# Posterior parameters
alpha = W + 1
beta_param = L + 1

# Draw posterior samples
n_samples = 100_000
p_samples = np.random.beta(alpha, beta_param, size=n_samples)
````

````python
#Visualizing the posterior distribution
x = np.linspace(0, 1, 200)
plt.plot(x, beta.pdf(x, alpha, beta_param))
plt.xlabel("p (probability of water)")
plt.ylabel("Density")
plt.title("Posterior Distribution: Beta(4, 12)")
plt.show()
````

````python
#Posterior predictive distribution
# Simulating 5 tosses
# Count how many times W appears

# Number of future tosses
n_future = 5

# Simulate posterior predictive outcomes
water_future = binom.rvs(
    n=n_future,
    p=p_samples
)
````

````python
# Count frequencies
values, counts = np.unique(water_future, return_counts=True)

# Convert to probabilities
predictive_probs = counts / counts.sum()

for v, p in zip(values, predictive_probs):
    print(f"P({v} waters in 5 tosses) ≈ {p:.3f}")
````
